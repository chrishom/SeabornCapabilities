```python
tips = sns.load_dataset('tips')
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False, bins = 50)
```

```python
sns.jointplot(x = tips['total_bill'], y = tips['tip'], data = tips, kind = 'kde')
```

```python
sns.jointplot(x = tips['total_bill'], y = tips['tip'], data = tips, kind = 'hex')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.rugplot(tips['total_bill'], color = 'r')
sns.distplot(tips['total_bill'], kde = False)
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
sns.distplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['tip'])
sns.rugplot(tips['tip'])
sns.distplot(tips['tip'])
```
